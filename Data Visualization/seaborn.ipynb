```python
import pandas as pd
import seaborn as sns
```

```python
df = sns.load_dataset("titanic")
df
```

```python
#histogram of age distribution
fig = sns.histplot(df, x="age", kde=True, color="blue", edgecolor="black")
fig.set_title("Age Distribution of Titanic Passengers")
fig.set_xlabel("Age")
fig.set_ylabel("Frequency")
```

```python
#countplot : survival based on gender
fig = sns.countplot(df, x ='survived')
fig.set_title("Survival Count of Titanic Passengers")
fig.set_xlabel("Survived")
fig.set_ylabel("Count")
```

```python
fig = sns.barplot(df, x='sex', y='fare', hue='class')
fig.set_title("Average Fare by Passenger Class and Gender")
fig.set_xlabel("sex")
fig.set_ylabel("Average Fare")
```

```python
fig = sns.countplot(df, x = 'class', palette = 'viridis')
fig.set_title("Count of Passengers by Class")
fig.set_xlabel("Passenger Class")
fig.set_ylabel("Count")
```

```python
sns.pairplot(df[['age','fare','survived','pclass']], hue='survived', palette='terrain')
```
